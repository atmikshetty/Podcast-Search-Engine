## The Python code below is for Transcript collection from the video_ids collected by Youtube Data API


```python
# Accessing the CSV Files
import pandas as pd

df1 = pd.read_csv('/content/joe_rogan.csv')
df2 = pd.read_csv('/content/chris_williamson.csv')
df3 = pd.read_csv('/content/diary_of_a_ceo.csv')
```

```python
# Merging the CSVs

df = pd.concat([df1,df2,df3])
df.size
```

```python
# Converting the dataframe to a list to pass to the get_transcript()

video_ids = df['id'].tolist()
len(video_ids)
```

```python
# A list to hold all the transcripts

transcripts = []
```

```python
# Generating transcripts

for id in video_ids:
  try:
    transcript = YouTubeTranscriptApi.get_transcript(id)
    transcripts.append({"id": id, "transcripts": transcript})
  except Exception as e:
    print(f"Could not get the transcript for {id}")
```

```python
len(transcripts)
```

```python
transcripts_df = pd.DataFrame(transcripts)
```

```python
transcripts_df.head()
```

```python
transcripts_df.to_csv('transcripts.csv', index=False)
```
